```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
impute = SimpleImputer(missing_values = np.nan,strategy = 'mean')
replaced_data = impute.fit_transform([
    [1,2],
    [np.nan,3],
    [4,5]
])
print(replaced_data)
```

```python
impute = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
replaced_data = impute.fit_transform([
    [2,2],
    [np.nan,2],
    [2,5],
    [3,6]
])
print(replaced_data)
```

```python
impute = SimpleImputer(missing_values = 3,strategy = 'most_frequent')
replaced_data = impute.fit_transform([
    [2,2],
    [1,2],
    [2,5],
    [3,6]
])
print(replaced_data)
```

```python
impute = SimpleImputer(missing_values = 3,strategy = 'constant')
replaced_data = impute.fit_transform([
    [2,2],
    [1,2],
    [2,5],
    [3,6]
])
print(replaced_data)
```

```python
df = pd.read_csv('Bengaluru_House_Data.xls').sort_values(by = ['price'])
plt.boxplot(df['price'])
plt.show()
```

```python
df = pd.read_csv('Bengaluru_House_Data.xls').sort_values(by = ['price'])
df['price_zscore'] = stats.zscore(df['price'])
plt.boxplot(df['price_zscore'])
plt.show()
```

```python
df = pd.read_csv('Bengaluru_House_Data.xls').sort_values(by = ['price'])
df['price_zscore'] = stats.zscore(df['price'])
df =df[df['price_zscore'] <= 3]
df =df[df['price_zscore'] >= -3]
plt.boxplot(df['price_zscore'])
plt.show()
```

```python
df = pd.read_csv('Bengaluru_House_Data.xls')
df = pd.read_csv('Bengaluru_House_Data.xls').sort_values(by = ['bath'])
df = df.fillna(value = 0)
plt.boxplot(df['bath'])
plt.show()
```

```python
df = pd.read_csv('Bengaluru_House_Data.xls').sort_values(by = ['bath'])
df = df.fillna(value = 0)
df['bath_zscore'] = stats.zscore(df['bath'])
df =df[df['bath_zscore'] <= 3]
df =df[df['bath_zscore'] >= -3]
plt.boxplot(df['bath_zscore'])
plt.show()
```

```python
# outlier another method (percentiles)
#percentile = no of values below x/ total no of values
# to take only 25% to 75% value or 0.25 to 0.75
```

```python
df = pd.read_csv('Bengaluru_House_Data.xls').sort_values(by = ['price'])
df = df.fillna(value = 0)
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
df = df[df['price'] > Q1]
df = df[df['price']< Q3]
plt.boxplot(df['price'])
plt.show()
```
