###### import necessary libraries

```python
import numpy as np
import pandas as pd
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
college1 = pd.read_csv('https://raw.githubusercontent.com/sethukrish6420/sethu/main/college_1.csv')
college2 = pd.read_csv('https://raw.githubusercontent.com/sethukrish6420/sethu/main/college_2.csv')
college = college1.merge(college2,how='outer')
print(college.shape)
college.head()
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
exceed1 = college[college['CodeKata Score']>15000]
exceed1.to_csv("Exceeded expectations.csv")
```

```python
exceed2 = college[(college['CodeKata Score']>10000) & (college['CodeKata Score'] < 15000)]
exceed2.to_csv("Reached_expectations.csv")
```

```python
exceed3 = college[(college['CodeKata Score']>7000) & (college['CodeKata Score'] < 10000)]
exceed3.to_csv("Needs_Improvement.csv")
```

```python
exceed4 = college[ college['CodeKata Score'] < 7000]
exceed4.to_csv("Unsatisfactory.csv")
```

###### Department wise codekata performence (pie chart)

```python
import matplotlib.pyplot as plt

codekata = college.groupby(college['Department'])['CodeKata Score'].sum()
plt.pie(codekata,labels=codekata.index,autopct='%.1f%%')
plt.show()
```
